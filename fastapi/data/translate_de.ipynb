```python
textfile = open("de_words_temp.txt", "w")

for line in lines:
    code = line.replace(" ", "_").split()[0]
    textfile.write(code + "\t")
    
    words = line.replace(" ", "_").split()[1:]
    for word in words:
        word = word.replace("_", " ")
        de_word = GoogleTranslator(source='en', target='de').translate(word) 
        textfile.write(de_word + ",")
    textfile.write("\n")
textfile.close()
    
```
